```python
import pandas as pd

# Load the data assuming it's in a single-column format with tab-separated values
# Replace 'your_file.txt' with your actual file name
df = pd.read_csv('C:/Users/nanduchowdary/Downloads_file.csv/marketing_campaign.csv', sep="\t", engine="python")

# Display the first few rows
print(df)
```

```python
print(df.isnull())
```

```python
df=df.drop_duplicates()
print(df)
```

```python
# ----- Standardize Education -----
df['Education'] = df['Education'].str.strip().str.title()
df['Education'] = df['Education'].replace({
    '2N Cycle': 'Second Cycle',
    'Basic': 'Basic',
    'Graduation': 'Graduate',
    'Master': 'Masters',
    'Phd': 'PhD'
})

# ----- Standardize Marital_Status -----
df['Marital_Status'] = df['Marital_Status'].str.strip().str.title()
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'Widow': 'Widowed',
    'Yolo': 'Single',
    'Together': 'Living Together',
    'Divorced': 'Divorced',
    'Married': 'Married',
    'Single': 'Single'
})

# ----- Standardize Gender (if present) -----
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].str.upper().replace({
        'M': 'Male',
        'F': 'Female',
        'MALE': 'Male',
        'FEMALE': 'Female'
    })

# ----- Standardize Country (if present) -----
if 'Country' in df.columns:
    df['Country'] = df['Country'].str.strip().str.title().replace({
        'Usa': 'USA',
        'Uk': 'UK',
        'Uae': 'UAE',
        'United States': 'USA',
        'United Kingdom': 'UK',
        'United Arab Emirates': 'UAE'
    })

# ----- View standardized values -----
print(df[['Education', 'Marital_Status']])
```

```python
# Convert 'Dt_Customer' to datetime object
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors='coerce')

# Format all dates to 'dd-mm-yyyy' as strings
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%d-%m-%Y')

# Check result
print(df['Dt_Customer'])
```

```python
# Clean and rename all column headers
df.columns = (
    df.columns.str.strip()             # remove leading/trailing whitespace
              .str.lower()             # convert to lowercase
              .str.replace(' ', '_')   # replace spaces with underscores
              .str.replace(r'[^\w_]', '', regex=True)  # remove non-alphanumeric characters
)

# Preview the new column names
print(df.columns)
```

```python
import pandas as pd
from datetime import datetime

# ----- Step 1: Load the CSV -----
file_path = r"C:\Users\nanduchowdary\Downloads_file.csv\marketing_campaign.csv"
df = pd.read_csv(file_path, sep='\t')  # Try sep=',' if tab fails

# ----- Step 2: Clean column names -----
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace(r'[^\w_]', '', regex=True)
)

# ----- Step 3: Fix Dt_Customer column -----
# Check exact name after cleanup
if 'dt_customer' in df.columns:
    df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True, errors='coerce')
else:
    raise KeyError("The column 'Dt_Customer' was not found. Check the column names.")

# ----- Step 4: Fix numeric types -----
numeric_columns = [
    'year_birth', 'income', 'kidhome', 'teenhome', 'recency',
    'mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts',
    'mntsweetproducts', 'mntgoldprods', 'numdealspurchases',
    'numwebpurchases', 'numcatalogpurchases', 'numstorepurchases',
    'numwebvisitsmonth', 'acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3',
    'acceptedcmp4', 'acceptedcmp5', 'complain', 'z_costcontact',
    'z_revenue', 'response'
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce', downcast='integer')

# ----- Step 5: Add 'age' column -----
df['age'] = datetime.now().year - df['year_birth']
df['age'] = df['age'].astype('Int64')  # handles NaNs gracefully

# ----- Step 6: Preview result -----
print(df[['dt_customer', 'year_birth', 'age']].head())
print("\nData types:\n", df.dtypes)

# ----- Step 7: Save cleaned dataset -----
df.to_csv("marketing_campaign_cleaned.csv", index=False)
```
